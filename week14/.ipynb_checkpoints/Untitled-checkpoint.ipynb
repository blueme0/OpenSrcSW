```python
# 실습문제 1번

import numpy as np
import csv
import matplotlib.pyplot as plt

f = open('seoul_people.csv', encoding='cp949')
data = csv.reader(f)
header = next(data)
data = list(data)

name = "광진구"
min_name = ""
max_name = ""

for row in data:
    if name in row[0]:
        temp = row
        for i in range(len(temp)):
            temp[i] = temp[i].replace(",","")
        home = np.array(temp[3:], dtype=int)
        break

isFirst = True
isFirstNum = True

for row in data:
    if not isFirstNum:
        temp = row
        for i in range(len(temp)):
            temp[i] = temp[i].replace(",","")
        away = np.array(temp[3:], dtype=int)
        s = np.sum((home-away)**2)
        tempS = round(np.sqrt(s/100), 2)
        tempT = str(tempS)
        if name not in row[0]:
            if isFirst:
                mn = s
                mx = s
                min_name = row[0].split()[1] + " (" + tempT + ")"
                min_result = away
                max_name = row[0].split()[1] + " (" + tempT + ")"
                max_result = away
                isFirst = False
            if s < mn:
                mn = s
                min_name = row[0].split()[1] + " (" + tempT + ")"
                min_result = away
            elif s > mx:
                mx = s
                max_name = row[0].split()[1] + " (" + tempT + ")"
                max_result = away
    isFirstNum = False
            
plt.style.use('ggplot')
plt.rc('font', family='AppleGothic')
plt.rc('axes', unicode_minus=False)
plt.title(name+' 지역과 가장 비슷한 인구 구조를 가진 지역')
plt.plot(home, label=name)
plt.plot(max_result, label=max_name)
plt.plot(min_result, label=min_name)
plt.legend()
plt.show()
```
