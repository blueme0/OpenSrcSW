```python
# 실습문제 1번

import csv
import matplotlib.pyplot as plt

f = open('seoul.csv', encoding='cp949')
data = csv.reader(f)
header = next(data)
high = []
low = []
for row in data:
    if len(row) > 0:
        if row[-1] != '':
            if row[0].split('-')[1] == '03' and row[0].split('-')[2] == '04':
                high.append(float(row[-1]))
                low.append(float(row[-2]))
f.close()
plt.plot(high, color='deeppink')
plt.plot(low, color='deepskyblue')
plt.show()
```

```python
# 실습문제 2번

import csv
import matplotlib.pyplot as plt

f = open('seoul.csv', encoding='cp949')
data = csv.reader(f)
header = next(data)
jan = []
aug = []
for row in data:
    if len(row) > 0:
        month = row[0].split('-')[1]
        if row[-1] != '':
            if month == '01':
                jan.append(float(row[-1]))
            elif month == '08':
                aug.append(float(row[-1]))
f.close()
plt.hist(aug, bins=100, color='r', label='Aug')
plt.hist(jan, bins=100, color='b', label='Jan')
plt.legend()
plt.show()
```
